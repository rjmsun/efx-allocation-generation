# EFX Graph Simulation Notebook
This notebook allows you to test and visualize graph structures such as the Champion Graph and Strong Envy Graph using random valuation distributions.

```python
# Imports
import sys
import os
# Add parent directory to Python path so we can import efx.py and graphs.py
sys.path.append(os.path.abspath('..'))
import importlib
import efx
importlib.reload(efx)
from efx import Allocation
from graphs import generate_val_distribution, build_champion_graph, build_strong_envy_graph, visualize_graph
import random
```

```python
# Parameters
num_agents = 4
num_items = 5
distribution = 'uniform'  # options: 'uniform', 'binary', 'sparse', 'correlated'
seed = 42
```

```python
# Generate random valuation matrix
valuations = generate_val_distribution(num_agents, num_items, distribution=distribution, seed=seed)
print("Valuations:")
for i, row in enumerate(valuations):
    print(f"Agent {i}: {row}")
```

```python
# Create random allocation (each item to one agent)
bundles = [[] for _ in range(num_agents)]
for item in range(num_items):
    bundles[random.randint(0, num_agents-1)].append(item)
allocation = Allocation(bundles, valuations)
print("\nAllocation:", allocation.bundles)
```

```python
# Champion Graph
champion_edges = build_champion_graph(allocation)
visualize_graph(champion_edges, title="Champion Graph")
```

```python
# Strong Envy Graph
strong_envy_edges = build_strong_envy_graph(allocation)
visualize_graph(strong_envy_edges, title="Strong Envy Graph")
```
