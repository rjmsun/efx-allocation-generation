```python
# Parameters
num_agents = 4
num_items = 10
distribution = 'uniform'  # options: 'uniform', 'binary', 'sparse', 'correlated'
seed = 42

# Generate random valuation matrix
valuations = np.array([
    [ 5, 17, 10,  6, 89,  1,  4, 19, 17, 16],
    [16,  9, 11, 94,  7, 12, 10,  2,  2,  2],
    [85, 18,  8, 85, 96, 10,  6, 16,  7,  1],
    [85,  2,  8,  9, 13, 10,  2,  8,  9, 11]
])
print("Valuations:")
for i, row in enumerate(valuations):
    print(f"Agent {i}: {row}")
```
