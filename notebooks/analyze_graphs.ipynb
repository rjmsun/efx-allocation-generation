```python
# Imports
import sys
import os
# Add parent directory to Python path so we can import efx.py and graphs.py
sys.path.append(os.path.abspath('..'))
import importlib
import efx
importlib.reload(efx)
from efx import Allocation
from graphs import *
import random
import numpy as np
```

```python
# Parameters
num_agents = 4
num_items = 8
distribution = 'uniform'  # options: 'uniform', 'binary', 'sparse', 'correlated'
seed = 42

# Generate random valuation matrix
valuations = np.array([
   [ 9,  5, 17, 10, 14,  3,  3, 99],
   [ 8,  2, 13,  9, 18, 19, 13, 16],
   [82,  3, 19,  6,  1,  8,  2,  8],
   [80,  3,  9, 18, 13, 12, 11, 100]
])
print("Valuations:")
for i, row in enumerate(valuations):
    print(f"Agent {i}: {row}")
```

```python
# Champion Graph
champion_edges = build_champion_graph(allocation)
visualize_graph(champion_edges, title="Champion Graph")
```
